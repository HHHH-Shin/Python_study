# Introduction 
---
### 프로그래밍 언어
- 저급 언어(low-level language): 기계어, 어셈블리 언어
- 고급 언어(high-level language): C, C++, Java, Python ...

### 컴파일러 vs. 인터프리터
컴파일러와 인터프리터는 작성된 프로그램을 기계어(0과1)바꿔 주는 소프트웨어.

고급 언어 프로그램(소스 프로그램) ➢ 컴파일러 or 인터프리터 ➢ 기계어 프로그램(목적 프로그램) ➢ 로더 ➢ 메모리어 기계어 프로그램 로드 ➢ 실행 ➢ 결과값

||컴파일러|인터프리터|
|:---:|:--:|:--:|
|번역방법|프로그램 전체 변역|실행되는 행 단위로 번역|
|장점|한번 번역하면 빠른 시간 내에 전체 실행 가능|큰 기억 장치가 필요하지 않으며 번역과정이 비교적 간단|
|단점|프로그램의 일부를 수정하는 경우에도 전체를 다시 컴파일|반복문이 많은 경우 매 반복 때마다 번역해야함|
|결과물|목적 언어로 된 프로그램|실행의 결과|
|예|FORTRAN, Pascal COBOL, Ada, C|Lisp, Scratch, Python|

### 파이썬 특징
- 실행 시간에 자료형을 검사
- 객체의 멤버에 무제한으로 접근할 수 있음
- 모듈, 클래스, 객체와 같은 언어의 요소가 내부에서 접근할 수 있음
- 리플렉션을 이용한 기술을 쓸 수 있음

|특징||
|:---:|:---:|
|하이브리드 언어|Java의 객체지향과 C의 함수지향의 특징을 동시에 갖는 대화형 인터프리터 언어|
|동적인 자료형 결정|동적으로 자료형을 결정. i.e. 자료형 선언이 없음|
|플랫폼 독립적 언어|대부분의 운영체제(윈도우, 리눅스, 유닉스 등)에서 동작|
|간경하고 쉬운 문법|비교적 간결하고, 쉬운 문법으로 프로그래밍 구현이 가능|
|풍부한 라이브러리와 오픈 소스|수 많은 라이브러리를 제공하며, 대부분 오픈 소스로 제공|
|메모리 자동 관리|쓰레기 수집(Garbage Collection) 기능을 이용하여 사용이 종료된 데이터를 자동으로 제거하여 메모리의 효율성을 높임|
